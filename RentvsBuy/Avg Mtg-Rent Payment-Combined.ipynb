```python

#make function to convert sales price to monthly payments

def sale_to_rent(sale):
    end = 0
    months = 360
    interest = .04
    Principal = sale - end
    pay_a = (interest / 12) / (1 - (1+interest/12)**(-months)) * Principal
    pay_b = interest / 12 * end
    monthly_payment = (interest / 12) * (1 / (1 - (1+interest/12) ** (-months))) * Principal + end
                   
    
    return (monthly_payment)



```

```python

range2=population_df2.loc[:, '2016-10':'2019-01']
range2.head()
#range2=range2.fillna(method='ffill')


```

```python
range2.loc[:, '2016-10':'2019-01'] = range2.loc[:, '2016-10':'2019-01'].applymap(sale_to_rent)
range2.head()
```

```python
#add index back on the header

dfrest_rent=df_city_rent.reset_index()
dfrest_mtg=df_city.reset_index()


```

```python
#combined rent df and mtg df
merged_df=pd.merge(dfrest_mtg, dfrest_rent, left_on='index', right_on='index')

#averaged each city
avg_merged=merged_df.mean()

#sorted the cities so they would be next to each other
sortedavgmerged=avg_merged.reindex(sorted(avg_merged.index), axis=1)

sortedavgmerged
```

```python
#plot map
plt.title('Average Mtg vs Rent Monthly Payment by City', fontsize=10)
plt.xlabel('Cities (_x=mtg, _y=rent)')
plt.ylabel('Monthly Payment')
plt.tight_layout()
sortedavgmerged.plot(kind='bar',grid=True) 

#save graph and show
plt.savefig(f'fig/combinedmtg_vs_rent.png', bbox_inches='tight')
plt.show()
```
