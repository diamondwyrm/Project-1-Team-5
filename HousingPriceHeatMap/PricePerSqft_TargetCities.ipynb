```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read in data set (data taken from Zillow Research)
file = 'data/sqft.csv'
df_raw = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
#define target cities
regions = ['New York, NY', 'Los Angeles-Long Beach-Anaheim, CA','Chicago, IL','Dallas-Fort Worth, TX',
           'Philadelphia, PA', 'Miami-Fort Lauderdale, FL','Atlanta, GA', 'Cleveland, OH',
           'Seattle, WA', 'Phoenix, AZ']

#set index to target cities
df1 = df_raw.set_index('RegionName')

#remove columns not needed
df=df1.drop(columns=['RegionID', 'SizeRank'])
```

```python
#create dictionary to store target city data
city_dict= {}

#store target city data
for region in regions:
    city_dict[region] = df.loc[region]

#convert to dataframe
df_city = pd.DataFrame(city_dict)

#display created dataframe
df_city
```

```python
#define tick values for graph
values = list(df.columns)
major_ticks = [value for value in values[::12]]

#define x-axis values
x_max = len(df_city)
x_min = 0
```

```python
#define graph parameters
figsize=(50,30)
fontsize=50
linewidth=10

#initialize figure
fig, ax = plt.subplots(figsize=figsize)

#plot all cities on graph
for i in df_city:
    ax.plot(df_city[i], lw=linewidth)
    ax.set_xticks(major_ticks, minor=False)
    
#graph formatting
plt.title('Median Home Value per Sqt Foot', fontsize=fontsize)
plt.ylabel('Price Per Sq Ft', fontsize=fontsize)
plt.xlabel('Month', fontsize=fontsize)
plt.tick_params(axis='both', which='major', labelsize=50)
plt.xticks(rotation=45)
plt.grid(which='major')

#legend formatting
leg = plt.legend(loc='center left', bbox_to_anchor=(1, .5), title='Cities', ncol=1, fontsize=50)
leg.set_title('City',prop={'size':fontsize})

#save graph and show
plt.savefig(f'figures/MedianSqFt.png', bbox_inches='tight')
plt.show()
```
